# ðŸ“Š Matplotlib Basics for Data Visualization


**Matplotlib** is a popular Python library for creating visualizations like **line plots**, **bar charts**, **scatter plots**, and more.
It is often used with **NumPy** and **Pandas** for data visualization in **Machine Learning** and **Data Science**.


```python
import matplotlib.pyplot as plt
import numpy as np
```

## ðŸ“ˆ 1. Line Plot
#### Trends, continuous data

```python

x = np.linspace( 0, 10, 100)
y = np.cos(x)

plt.plot(x, y)
plt.title("cosine Wave")
plt.xlabel("x")
plt.ylabel("cos(x)")
plt.grid(True)
plt.show()
```

## ðŸ“Š 2. Bar Chart
### Categorical comparisons

```python

categories = ['A', 'B', 'C', 'D', 'E'] 
values = [3, 7, 5, 6,20]

plt.bar(categories, values, color='orange')
plt.title("Bar Chart Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()


```

## âš« 3. Scatter Plot
### Relationship, randomness
A scatter plot displays individual data points as dots on a 2D plane.

Shows the relationship (correlation) between two variables.
Visualizing correlation between variables (e.g., height vs weight).

Detecting clusters, trends, or outliers.

Used in regression analysis, clustering, and EDA (Exploratory Data Analysis).

```python

x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y, color='green')
plt.title("Random Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

## ðŸ“Š 4. Histogram
 ### Data distribution, A histogram shows the distribution of a dataset.

It groups values into intervals (bins) and displays the frequency of data points in each bin.

Each bar represents the number of occurrences of data points in a particular range.
Analyzing data distribution (e.g., student scores, heights, stock returns).

Identifying skewness, kurtosis, or outliers in data.

Evaluating probability distributions in statistics or ML.

```python

data = np.random.randn(1000)
plt.hist(data, bins=30, color='purple', edgecolor='black')
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

```

## ðŸ¥§ 5. Pie Chart
### A pie chart represents proportional data in a circular format.

Each sliceâ€™s size corresponds to a percentage of the whole.
Showing market share, budget allocation, survey results, etc.

Best for categorical data with limited categories.

Not ideal if too many categories or if values are close together.

```python

labels = ['Python', 'Java', 'C++', 'Ruby']
sizes = [40, 30, 20, 10]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title("Language Popularity")
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

## ðŸŽ¨ 6. Customizing Plots (Color, Style, Grid)

Helps to enhance visual appeal and readability of your plots.

You can change colors, line styles, add legends, grids, and labels.
Used in comparison plots like multiple lines on a graph.

Enhances clarity and presentation for dashboards, reports, and interactive apps.

Essential for multi-line graphs, scientific charts, and custom dashboards.

```python

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label='sin(x)', linestyle='--', color='blue')
plt.plot(x, y2, label='cos(x)', linestyle='-.', color='red')
plt.title("Sin & Cos Waves")
plt.xlabel("x")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.show()
```
