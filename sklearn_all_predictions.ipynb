# ü§ñ Scikit-learn Prediction Examples


This notebook demonstrates how to use **Scikit-learn** and **Pandas** to build predictive models from real-life scenarios:

1. üìò Student Marks Prediction based on Study Hours  
2. ‚õΩ Vehicle Fuel Expense Prediction based on Fuel Consumed  
3. üåßÔ∏è Rainfall Prediction based on Temperature and Humidity (with messy data cleaned)


## üìò Student Marks Prediction

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Dataset: Hours studied vs Marks scored
data1 = {
    # 'Hours': [1, 2, 3, 4, 5, 6, 7, 8, 9],
    # 'Marks': [10, 20, 30, 40, 50, 60, 65, 78, 88]
    'Hours': [1, 2, 3, 4, 5, 6, 7, 8, 9],
    'Marks': [10, 20, 30, 40, 50, 60, 70, 80, 90] 
}
df1 = pd.DataFrame(data1)

# Plot the data
plt.scatter(df1['Hours'], df1['Marks'], color='blue')
plt.title("Hours vs Marks")
plt.xlabel("Hours Studied")
plt.ylabel("Marks Scored")
plt.grid(True)
plt.show()

# Prepare and split the data
X1 = df1[['Hours']]
y1 = df1['Marks']
X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, y1, test_size=0.2, random_state=1)

# Train model
model1 = LinearRegression()
model1.fit(X_train1, y_train1)
predictions1 = model1.predict(X_test1)

# Evaluation and prediction
print("Predictions:", predictions1)
print("Mean Squared Error:", mean_squared_error(y_test1, predictions1))

# Predict for 7.5 hours
hours = 7.5
predicted_marks = model1.predict([[hours]])
print(f"Predicted marks for {hours} hours studied: {predicted_marks[0]:.2f}")

# Visualize regression line
plt.scatter(X1, y1, color='blue')
plt.plot(X1, model1.predict(X1), color='red')
plt.title("Regression Line: Hours vs Marks")
plt.xlabel("Hours Studied")
plt.ylabel("Marks Scored")
plt.grid(True)
plt.show()
```

## ‚õΩ Vehicle Fuel Expense Prediction

```python

# Dataset: Fuel consumed (in litres) vs Expense (in INR)
data2 = {
    # 'Litres': [5, 10, 15, 20, 25, 30, 35],
    # 'Expense': [500, 1100, 1500, 2200, 2500, 3300, 3500]  #18L - 1869
    'Litres': [5, 10, 15, 20, 25, 30, 35],
    'Expense': [510, 1020, 1530, 2040, 2550, 3060, 3570]  
}
df2 = pd.DataFrame(data2)

# Plot the data
plt.scatter(df2['Litres'], df2['Expense'], color='green')
plt.title("Fuel Consumed vs Expense")
plt.xlabel("Fuel Consumed (Litres)")
plt.ylabel("Expense (INR)")
plt.grid(True)
plt.show()

# Prepare data
X2 = df2[['Litres']]
y2 = df2['Expense']
X_train2, X_test2, y_train2, y_test2 = train_test_split(X2, y2, test_size=0.3, random_state=42)

# Train model
model2 = LinearRegression()
model2.fit(X_train2, y_train2)
predictions2 = model2.predict(X_test2)

# Evaluation
print("Predictions:", predictions2)
print("Mean Squared Error:", mean_squared_error(y_test2, predictions2))

# Predict for 18 litres
litres = 19
predicted_expense = model2.predict([[litres]])
print(f"Predicted expense for {litres} litres of fuel: ‚Çπ{predicted_expense[0]:.2f}")

# Visualize regression line
plt.scatter(X2, y2, color='green')
plt.plot(X2, model2.predict(X2), color='red')
plt.title("Regression Line: Fuel Consumption vs Expense")
plt.xlabel("Fuel Consumed (Litres)")
plt.ylabel("Expense (INR)")
plt.grid(True)
plt.show()
```

## üåßÔ∏è Rainfall Prediction with Cleaned Data

```python

# Simulated messy dataset
data3 = {
    'Temperature': [30, 28, '29', 35, np.nan, 33, '31', 32, 30, ' '],
    'Humidity': [70, 65, 60, 80, 75, ' ', 68, np.nan, 72, 70],
    'Rainfall': [100, 85, '90', 120, 110, 95, 'NaN', 105, 102, '']
}
df3 = pd.DataFrame(data3)

# Clean the data
df3.replace([' ', 'NaN', ''], np.nan, inplace=True)
df3['Temperature'] = pd.to_numeric(df3['Temperature'], errors='coerce')
df3['Humidity'] = pd.to_numeric(df3['Humidity'], errors='coerce')
df3['Rainfall'] = pd.to_numeric(df3['Rainfall'], errors='coerce')
df3.dropna(inplace=True)

# Show cleaned data
print("Cleaned Data:\n", df3)

# Features and target
X3 = df3[['Temperature', 'Humidity']]
y3 = df3['Rainfall']
X_train3, X_test3, y_train3, y_test3 = train_test_split(X3, y3, test_size=0.2, random_state=42)

# Train model
model3 = LinearRegression()
model3.fit(X_train3, y_train3)
predictions3 = model3.predict(X_test3)

# Evaluate and predict
print("\nPredictions:", predictions3)
print("Mean Squared Error:", mean_squared_error(y_test3, predictions3))

# Predict for new value
new_data = pd.DataFrame({'Temperature': [31], 'Humidity': [66]})
predicted_rainfall = model3.predict(new_data)
print(f"Predicted Rainfall for Temp=31¬∞C & Humidity=66%: {predicted_rainfall[0]:.2f} mm")

# Plot: Rainfall vs Temperature
plt.figure(figsize=(8,5))
plt.scatter(df3['Temperature'], df3['Rainfall'], color='blue', label='Data Points')
plt.xlabel("Temperature (¬∞C)")
plt.ylabel("Rainfall (mm)")
plt.title("Rainfall vs Temperature")
plt.grid(True)
plt.legend()
plt.show()

# Plot: Actual vs Predicted
plt.figure(figsize=(8,5))
plt.scatter(y_test3, predictions3, color='purple', label='Predicted vs Actual')
plt.plot([y_test3.min(), y_test3.max()], [y_test3.min(), y_test3.max()], color='red', linestyle='--')
plt.xlabel("Actual Rainfall")
plt.ylabel("Predicted Rainfall")
plt.title("Actual vs Predicted Rainfall")
plt.grid(True)
plt.legend()
plt.show()
```
